Importamos las librerias necesarias e inicializamos el grafo

```python
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS

g = Graph()

EX = Namespace("http://abandoned.cars.com/")
RML = Namespace("http://semweb.mmlab.be/ns/rml#")
FOAF = Namespace("http://xmlns.com/foaf/0.1/")

g.namespace_manager.bind('ex', EX, override=False)
g.namespace_manager.bind('rml', RML, override=False)
g.namespace_manager.bind('foaf', FOAF, override=False)
g.namespace_manager.bind('rdf', RDF, override=False)
g.namespace_manager.bind('rdfs', RDFS, override=False)

g.parse(github_storage, format="turtle")
```

##Consulta 1
Consultar las clases del grafo

```python
query = """
  SELECT DISTINCT ?class
  WHERE {
    ?s a ?class .
  }
  """

# Ejecutar la consulta
qres = g.query(query)

# Imprimir los resultados
for row in qres:
    print(row)
```

##Consulta 2
Consultar las propiedades del grafo

```python
query = """
  SELECT DISTINCT ?property
  WHERE {
    ?s ?property ?o .
  }
  """
# Ejecutar la consulta
qres = g.query(query)

# Imprimir los resultados
for row in qres:
    print(row)
```

##Consulta 3
Consultar las tripletas que tiene el grafo, lo limitamos a 100 para que no muestre todos los datos.

```python
query = """
  SELECT ?s ?p ?o
  WHERE {
    ?s ?p ?o .
  } LIMIT 100
  """

# Ejecutar la consulta
qres = g.query(query)

# Imprimir los resultados
for row in qres:
    print(row)
```

##Consulta 4
Consultar los distritos cuya region es "S East"

```python
#Cambiar la URI, falta un / entre "com" y "hasRegion"

query = """
  SELECT ?DISTRICT
  WHERE {
    ?DISTRICT ex:hasRegion "Eastern" .
  }
  """

# Ejecutar la consulta
qres = g.query(query)

# Imprimir los resultados
for row in qres:
    print(row)
```

##Consulta 5
Mostrar la cantidad total de coches abandonados en el año 2003, para la region S East.

```python
query = """
  SELECT (SUM(xsd:integer(?amount)) AS ?totalAbandonedCars)
  WHERE {
      ?district ex:hasRegion "S East" .
      ?data ex:hasYear "2003" .
      ?data ex:hasAmount ?amount .
      ?data ex:hasDistrict ?district .
  }
"""

# Ejecutar la consulta
qres = g.query(query)

# Imprimir los resultados
for row in qres:
    print(row)
```

##Consulta 6
Consultar los distritos con mayor numero de coches abandonados en el año 2004

```python
query = """
SELECT ?district (SUM(xsd:integer(?amount)) AS ?totalAbandonedCars)
WHERE {
    ?data ex:hasYear "2004" .
    ?data ex:hasDistrict ?district .
    ?data ex:hasAmount ?amount .
}
GROUP BY ?district
ORDER BY DESC(?totalAbandonedCars)
"""

# Ejecutar la consulta
qres = g.query(query)

# Imprimir los resultados
for row in qres:
    print(f"District: {row.district}, Total Abandoned Cars: {row.totalAbandonedCars}")
```
